```python
employee = []
again = "y"
while again == "y":
    detail = {}
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    work = input("Enter Work Feild: ")
    skills = input("Enter 3 Skills sperated by comma(,): ").split(",")
    degree_title = input("Enter Degree Title: ")
    degree_major = input("Enter Major Subject: ")
    degree_end_date = input("Enter Completion Date: ")
    degree = {"title" : degree_title, "major" : degree_major, "completionDate" : degree_end_date}
    salary = int(input("Enter Your Salary"))
    detail = {"name" : name,
             "age" : age,
             "work" : work,
             "skills" : skills,
             "degree" : degree,
             "salary" : salary}
    employee.append(detail)
    correct = 1
    while correct == 1:
        again = input("Do you want to add another detail? (y/n)")
        if again == "y" or again == "n":
            correct = 0
        else:
            print("Please Enter Only \"y\" or \"n\". (y/n) \n")
            correct = 1
```
